```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

sns.set()
sns.set_style("whitegrid", {'axes.grid' : True})
sns.set_color_codes()
#sns.set_style("white")
sns.set_context("notebook", font_scale=1.0, rc={"lines.linewidth": 2.5, 'lines.markeredgewidth': 1., 'lines.markersize': 10})
# axis labels
font = {'family' : 'serif'}
mpl.rc('font', **font)
```

```python
df = pd.read_csv("hack2outputs.csv")
```

```python
df.head()
```

```python
hue_order=["Separable Kernel","Shared Memory", "Original Kernel"]
```

```python
g = sns.catplot(data=df[(df['Device Name'] == 'V100') & (df['Block Size'] == 32)], kind="bar", hue_order=hue_order, hue='Version', x ="Input Image", y = "Duration", ci=None)
axes = g.axes.flatten()
axes[0].set_ylabel("Duration (μs)")
g.savefig("input_images.png")
```

```python
g = sns.catplot(data=df[(df['Input Image'] == 'monarch') & (df['Block Size'] == 32)], kind="bar", hue_order=hue_order, hue='Version', x ="Device Name", y = "Duration", ci=None)
axes = g.axes.flatten()
axes[0].set_ylabel("Duration (μs)")
g.savefig("devices.png")
```

```python
g = sns.catplot(data=df[(df['Input Image'] == 'monarch') & (df['Device Name'] == "V100")], kind="bar", hue_order=hue_order, hue='Version', x ="Block Size", y = "Duration", ci=None)
axes = g.axes.flatten()
axes[0].set_ylabel("Duration (μs)")
g.savefig("block_size.png")
```
